```python
import pandas as pd
import numpy as np
```

### Excercise

```python
df = pd.read_csv('deliveries.csv')
df.head()
```

## Groupby

```python
df.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10).reset_index()
```

## kis bowler nay sab say zayada wicket li?

```python
top_bowler=df.groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(5).reset_index()
top_bowler
```

## kis batsman nay sab say zayada 6 maray?

```python
six=df[df['batsman_runs']==6]
six.groupby('batter')['batsman_runs'].count().sort_values(ascending=False).head(10).reset_index()
```

## kis batsman nay kis bowler ko sab say zayada runs maray?

```python
df.groupby(['batter','bowler'])['batsman_runs'].sum().sort_values(ascending=False).head(10).reset_index()
```

## Kis batsman nay kis bowler ko kitni balls may kitnay runs maray?

```python
df.groupby(['batter','bowler']).agg({'batsman_runs':'sum','ball':'count'}).sort_values(by='batsman_runs',ascending=False).head(10).reset_index()
```

## Check MIN ,MAX,Mean,Median,Mode,std

```python
import seaborn as sns
titanic=sns.load_dataset('Titanic')
titanic.sample(5)
```

```python
print('Minimum Age is :',titanic['age'].min())
print("------------------------------")
print('Maximum Age is :',titanic['age'].max())
print('---------------------------------')
print('Mean Age is :',titanic['age'].mean())
```

## Fill missing values

```python
titanic.isnull().sum()
```

### Ham sirf Age column ki missing values say fill karain gay

```python
titanic['age']=titanic['age'].fillna(titanic['age'].mean())
titanic.isnull().sum()
```
