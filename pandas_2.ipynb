## Concat DataFrame

```python
data1 = {
    'Student_ID': [1, 2, 3],
    'Name': ['Ali', 'Sara', 'Ahmed']
}
df1 = pd.DataFrame(data1)
df1
```

```python
data2 = {
    'Student_ID': [4, 5, 6],
    'Name': ['Zara', 'Bilal', 'Ayesha']
}
df2 = pd.DataFrame(data2)
df2
```

```python
new_dataframe=pd.concat([df1,df2])
new_dataframe.reset_index(inplace=True)
new_dataframe.drop(columns=['index'])
```

## Inner join

```python
data3 = {
    'StudentID': [1, 2, 3],
    'Name': ['Ali', 'Sara', 'Ahmed']
}

df3 = pd.DataFrame(data3)
df3
```

```python
# Creating the first DataFrame


# Creating the second DataFrame
data4 = {
    'Student_ID': [2, 3, 4],
    'Grade': ['A', 'B', 'C']
}
df4 = pd.DataFrame(data4)
df4
```

```python
pd.merge(df3,df4,on='Student_ID',how='inner')
```
